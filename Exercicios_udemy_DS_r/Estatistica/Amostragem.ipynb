# Amostra simples

Amostras de entre 0-1, 150 amostras, com propabilidaddes equivalentes

```r
amostra = sample(c(0,1), 150, replace = TRUE, prob = c(0.5,0.5))
```

```r
amostra
```

```r
summary(as.factor(amostra))
```

```r
amostra1 = sample(c(1:1000), 150, replace=FALSE) # amostras sem repetição
```

```r
amostra1
```

Exemplo com dados reais dataset Iris

```r
amostra2 = sample(c(0,1),150,replace=TRUE, prob=c(0.7,0.3))
summary(as.factor(amostra2))
```

```r
amostra_iris = iris[amostra2==1,]
amostra_iris # pegou 47 amostras aleatorias do dataset
```

```r
dim(amostra_iris) # 47 amostras com 5 colunas
```

# Amostra Estratificada , representação balanceada dos dados

```r
#install.packages("lpSolve")
#install.packages("sampling")
```

```r
#obs : retirar o '#' para instalar os pacotes e depois recolocar pois não é necessario rodar essa linha novamente
```

```r
library("sampling")
```

```r
amostra_est_iris = strata(data=iris, stratanames = c("Species"), size=c(rep(25,3)),method = "srswor")
# o metodo "srswor" gera amostras simples sem reposição, para amostras com reposição usar o "srswr"
# a amostragem foi feito com base na coluna "Species"
```

```r
summary(amostra_est_iris)
```
