main (legena principal), ylab e xlab legenda dos eixos x e y

```r
hist(trees$Height, main = "Dataset Trees", ylab = "Freguencia", xlab="Altura",col="gray")
```

Grafico com base na densidade

```r
densidade = density(trees$Height)
plot(densidade,main = "Dataset Trees", ylab = "Freguencia", xlab="Altura")
```

Graficos sobrepostos

```r
hist(trees$Height,main=NULL, xlab=NULL,ylab=NULL)
par(new=TRUE)
plot(densidade, col="red")
```

Grafico de dispersão

```r
plot(trees$Girth, trees$Volume,main="Arvore",ylab="Circuferencia",xlab="Volume",col="red",pch=10) # pch é os tipos de icones
```

Tabela de Graficos

```r
split.screen(figs=c(2,2))
screen(1)
plot(trees$Girth, trees$Volume, pch=16)
screen(2)
plot(trees$Girth, trees$Height, pch=16)
screen(3)
plot(trees$Height, trees$Volume, pch=16)
screen(4)
hist(trees$Volume, pch=16, main="Dataset Trees", ylab="Freguencia", xlab="Volume")
close.screen(all=TRUE)
```
