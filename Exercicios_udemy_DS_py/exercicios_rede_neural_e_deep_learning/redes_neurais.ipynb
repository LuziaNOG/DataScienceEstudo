```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from yellowbrick.classifier import ConfusionMatrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
base = datasets.load_iris() # carrega o dataset
```

```python
previsores = base.data # variaveis explanatorias
classes = base.target # classes ou respostas
```

```python
classe_dummy = np_utils.to_categorical(classes) # dummy é um tipo de variavel categorica especial, podendo ser 0 ou 1 para informar se um elemento pertence ao um grupo ou nao
```

```python
x_train, x_test, y_train, y_test = train_test_split(previsores,classe_dummy, test_size=0.3, random_state=0)
```

```python
# possuem 5 neuronios
modelo = Sequential()
modelo.add(Dense(units=5, input_dim=4)) #camada de entrada - input_dim é 4 porque é a qtd de previsores
modelo.add(Dense(units=4)) # camada oculta
modelo.add(Dense(units=3, activation = 'softmax')) # camada de saida 3 pq é a qtd de classes
```

Softmax é uma função muito utilizada com problemas que possuem mais de 2 classes , ele retorna a propabilidade de cada classe

```python
modelo.summary()
```

```python
modelo.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
modelo.fit(x_train, y_train, epochs=1000, validation_data=(x_test,y_test)) # obs a ultima val_accuracy é a mesma tx de acerto do final.
```

O modelo retorna propabilidade do elemto no grupo, o np.argmax retorna o grupo com maior propabilidade com base no index

```python
previsoes = np.argmax(modelo.predict(x_test) > 0.5, axis = 1) 
type(previsoes)
```

```python
y_matrix = [np.argmax(y) for y in y_test] # transformando novamente em um array (pq tava categorizado)
type (y_test)
```

```python
matrix_conf = confusion_matrix(y_matrix,previsoes) # matriz de confusão para ver os acertos e erros
matrix_conf
```

```python
tx_acerto = accuracy_score(y_matrix,previsoes) # por fim a taxa de acerto
tx_acerto
```

Obs : o teste do modelo e as previsoes utilizaram os mesmos dados, isso pode ter interferido na taxa de acerto
